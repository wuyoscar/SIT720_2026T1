**KNeighborsClassifier** is a classification algorithm that belongs to the family of lazy learning algorithms. It means that it does not attempt to construct a general internal model, but simply stores the training dataset. Classification is computed from a simple majority vote of the nearest neighbors of each point. [More details](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html)<br>

### iris datasets

```python
# Import data and modules
import pandas as pd
import numpy as np
from sklearn import datasets
iris = datasets.load_diabetes()

# We'll use the petal length and width only for this analysis
X = iris.data
y = iris.target

# Place the iris data into a pandas dataframe
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

# View the first 5 rows of the data
print(iris_df.head())
```

### feature relationship

```python
import seaborn as sns
sns.pairplot(iris_df)
```

### checking class balance

```python
from collections import Counter
import matplotlib.pyplot as plt
counts = Counter(y)
plt.pie([float(v) for v in counts.values()], labels=[float(k) for k in counts],
           autopct='%1.1f%%')
plt.show()
```

```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt

markers = ('s', 'x', 'o')
colors = ('red', 'blue', 'lightgreen')
cmap = ListedColormap(colors[:len(np.unique(y))])
for idx, cl in enumerate(np.unique(y)):
    plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
               c=cmap(idx), marker=markers[idx], label=cl)
plt.show()
```

### Train and test data spliting

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=0)

print('There are {} samples in the training set and {} samples in the test set'.format(
X_train.shape[0], X_test.shape[0]))
print()
```

### data scaling

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

sc.fit(X_train)

X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

print('After z-scorescaling on features:\n')
print(pd.DataFrame(X_train_std, columns=iris_df.columns).head())
```

### feature distribution with mean,$\mu= 0$

```python
import seaborn as sns
plt.figure()
iris_df.hist()
plt.suptitle("Before scaling")
pd.DataFrame(X_train_std, columns=iris_df.columns).hist()
plt.suptitle("After scaling")
plt.show()
```

### train and test model

```python
from sklearn.neighbors import KNeighborsClassifier

neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train_std, y_train)

print('The accuracy of the KNN classifier on training data is {:.2f} out of 1'.format(neigh.score(X_train_std, y_train)))

print('The accuracy of the KNN classifier on test data is {:.2f} out of 1'.format(neigh.score(X_test_std, y_test)))
```

### PCA to reduce the diment into 2

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
reduced_data_train=pca.fit_transform(X_train_std)
reduced_data_test=pca.transform(X_test_std)
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(reduced_data_train, y_train)
```

### visulizing SVM boundaries

```python
print(__doc__)
import warnings
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
def bnd(X, y, classifier,scor,n_neighbors,test_idx=None, resolution=0.02):
  
    h = .02  # step size in the mesh

    # we create an instance of SVM and fit out data. We do not scale our
    # data since we want to plot the support vectors
    C = 1.0  # SVM regularization parameter

   


    # Determine the maximum and minimum mesh as a boundary
    x_min, x_max = X[:, 0].min() - .1, X[:, 0].max() + .1
    y_min, y_max = X[:, 1].min() - .1, X[:, 1].max() + .1
    # Generating the points on the mesh
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                         np.linspace(y_min, y_max, 100))


    
    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, x_max]x[y_min, y_max].
    # plt.subplot(1, 1, 1)
    # plt.subplots_adjust(wspace=0.4, hspace=0.4)

    Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    fig=plt.figure(2)
    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
   
    # Plot also the training points
    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
                    alpha=0.8, c=cmap(idx),
                    marker=markers[idx], label=cl)
    plt.title('The accuracy of the KNN classifier on test data is {:.2f} out of 1 when N={}'.format(scor,n_neighbors))
    plt.xlabel('PCA-1')
    plt.ylabel('PCA-2')
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xticks(())
    plt.yticks(())

    plt.show()
```

# Decision tree

**Decision Trees (DTs)** are a non-parametric supervised learning method used for classification and regression. The goal is to create a model that predicts the value of a target variable by learning simple decision rules inferred from the data features. [More details](https://scikit-learn.org/stable/modules/tree.html#:~:text=Decision%20Trees%20(DTs)%20are%20a,as%20a%20piecewise%20constant%20approximation.)
